```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
daily_sales = daily_sales.set_index('date')
```

```python
daily_sales.head()
```

```python
train_size = int(len(daily_sales) * 0.8)

train = daily_sales[:train_size]
test = daily_sales[train_size:]
```

```python
print(len(train), len(test))
```

```python
model = ARIMA(train['sales'], order=(5,1,0))
model_fit = model.fit()
```

```python
print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=len(test))
```

```python
forecast.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.plot(train.index, train['sales'], label='Training')
plt.plot(test.index, test['sales'], label='Actual')
plt.plot(test.index, forecast, label='Predicted')
plt.legend()
plt.title("ARIMA Forecast")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(test['sales'], forecast))
print("RMSE:", rmse)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
data = daily_sales[['sales']].values

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data)
```

```python
sequence_length = 30

X = []
y = []

for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i-sequence_length:i, 0])
    y.append(scaled_data[i, 0])

X = np.array(X)
y = np.array(y)
```

```python
print(X.shape)
print(y.shape)
```

```python
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python
print(X.shape)
```

```python
train_size = int(len(X) * 0.8)

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]
```

```python
model = Sequential()

model.add(LSTM(50, return_sequences=False, input_shape=(X_train.shape[1],1)))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(
    X_train,
    y_train,
    batch_size=16,
    epochs=20,
    validation_data=(X_test, y_test)
)
```

```python
predictions = model.predict(X_test)
```

```python
predictions = scaler.inverse_transform(predictions.reshape(-1,1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
print("LSTM RMSE:", rmse)
```

```python
plt.figure(figsize=(15,5))
plt.plot(y_test_actual, label='Actual')
plt.plot(predictions, label='Predicted')
plt.legend()
plt.title("LSTM Forecast")
plt.show()
```

```python
print("LSTM RMSE:", rmse)
```
